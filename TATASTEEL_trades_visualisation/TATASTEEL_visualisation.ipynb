* Volume:- Volume of trade is the total quantity of shares or contracts traded for a     specified security.

# Using Matpllotlib to visualize the data

* Importing the necessary library for visualisation

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
```

* Plotting the statistical infromation of 4 columns which are:-
    * **Open**
    * **Close**
    * **Low**
    * **High**

```python
fig,ax=plt.subplots()
plt.plot(description.index,description["Open"],"ro--",markersize=20)
plt.plot(description.index,description["Close"],"bo--",markersize=15)
plt.plot(description.index,description["Low"],"go--",markersize=10)
plt.plot(description.index,description["High"],"co--", markersize=5)
plt.grid()
```

* Plotting each columns of dataset into graph .

```python
fig,ax=plt.subplots(11,1,figsize=(15,18))
fig.tight_layout(pad=2)

a=0
color = iter(["red","green","blue","lightgreen","hotpink","cyan","purple","crimson","black","grey","limegreen"])
for i in vis_data.columns:
    
    vis_data[i].plot(ax=ax[a],color=next(color))
    
    ax[a].legend()
    plt.margins()
    a+=1
```

* Here we can see that the trades started from mid of 2011

```python
fi,ax=plt.subplots(figsize=(10,5))
plt.grid()
plt.plot(vis_data["Trades"])
```

* visualizing the comparison between **open** and **close** price.

```python
fig,ax=plt.subplots(figsize=(10,5))

x=np.linspace(60,1055,5306)
c1=vis_data["Open"]
c2=vis_data["Close"]
ax.fill_between(x,c1,c2,where=c1>c2,alpha=1,color="red",hatch="|")
```

* Visualising the comaprison between **High** and **Low**.
    * WHere,
        **difference=(High-Low)
      &, difference>50**
      


```python
fig,ax=plt.subplots(figsize=(10,5))

x=np.linspace(60,1055,5306)
c1=vis_data["High"]
c2=vis_data["Low"]
ax.fill_between(x/2,c1,c2,where=(c1-c2)>50,alpha=1,color="limegreen",hatch="|")
```

* Here we can see that sometimes there was a rise of stock price more than 50.
